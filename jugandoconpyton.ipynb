```
from pylab import *
x0= 0
x1= 100.0 
dx = 0.1
n = int(ceil((x1-x0)/dx) +1 )
x = zeros((n,1),float)
y = zeros((n,1),float)
i = 0
while i < n:
    x[i] = x0 + i*dx
    y[i] = sin(x[i])
    i = i+1
plot(x,y)
show()
```
